```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

```

```python
# Define the column names
col_names = ['polarity', 'aspect_category', 'target_term', 'char_offset', 'sentence']

# Read the CSV file with the specified column names
dev = pd.read_csv('data\\devdata.csv', sep='\t', names=col_names)
train = pd.read_csv('data\\traindata.csv', sep='\t', names=col_names)
```

```python
# create two new columns for aspect category: category_root and category_sub

train['category_root'] = train['aspect_category'].apply(lambda x: x.split('#')[0])
train['category_sub'] = train['aspect_category'].apply(lambda x: x.split('#')[1])

dev['category_root'] = dev['aspect_category'].apply(lambda x: x.split('#')[0])
dev['category_sub'] = dev['aspect_category'].apply(lambda x: x.split('#')[1])

# change position the two new columns to be after the column aspect_category

train = train[['polarity', 'aspect_category', 'category_root', 'category_sub', 'target_term', 'char_offset', 'sentence']]

dev = dev[['polarity', 'aspect_category', 'category_root', 'category_sub', 'target_term', 'char_offset', 'sentence']]

# drop the column aspect_category since we have the two new columns category_root and category_sub
train = train.drop(['aspect_category'], axis=1)
dev = dev.drop(['aspect_category'], axis=1)
```
